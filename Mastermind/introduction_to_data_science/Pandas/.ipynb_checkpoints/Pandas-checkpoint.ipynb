# Funciones Avanzadas

```python
df_nuevo = df.copy()
```

```python
df_nuevo['nombre'] = ['Batman', 'Spiderman', 'Catwoman', 'Capitan America', 'Thor', 'Iron Man', 'Loki']
df_nuevo
```

```python
df
```

## df.where()

```python
filtro1 = df['especialidad'] == 'Programación'
filtro2 = df['visitas'] > 145000
```

```python
df.where(filtro1 & filtro2).dropna().loc[:,'nombre']
```
